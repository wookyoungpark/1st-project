```python
import numpy as np
from scipy import optimize, linalg
import matplotlib.pyplot as plt
from scipy.special import hermite, factorial 

# ============================================================
# 0️⃣ 기본 설정
# ============================================================
HBAR = 1.0
M = 1.0
KINETIC_FACTOR = - (HBAR**2) / (2.0 * M)

X_LIMIT = 6.0 
X_GRID = np.linspace(-X_LIMIT, X_LIMIT, 1000)
DX = X_GRID[1] - X_GRID[0]

plt.rcParams['axes.unicode_minus'] = False 


# ============================================================
# 1️⃣ 퍼텐셜 정의 (총 7개)
# ============================================================
def V_sho(x): 
    """예시 1: 조화 진동자 (검증용)"""
    return 0.5 * x**2

def V_anharmonic(x): 
    """예시 2: 비조화 진동자 (x^4)"""
    return 0.5 * x**2 + 0.1 * x**4

def V_double_well(x): 
    """예시 3: 이중 우물 (NH3 모델)"""
    return -10.0 * x**2 + 1.0 * x**4

def V_finite_well(x):
    """예시 4: 유한 사각 우물 """
    V0 = 20.0  # 우물의 깊이
    a = 1.0    # 우물의 절반 폭
    return np.where(np.abs(x) < a, -V0, 0.0)

def V_soft_coulomb_toy(x):
    """예시 5: 1D 소프트-쿨롱 (V=-1/r)"""
    a = 0.05  
    return -1.0 / np.sqrt(x**2 + a**2)

def V_hydrogen_p_orbital(x):
    """예시 6: 1D 수소 p-오비탈 (l=1)"""
    l = 1      
    Z = 1      
    a = 0.05   
    V_centrifugal = l * (l + 1) / (x**2 + a**2)
    V_coulomb = -2 * Z / np.sqrt(x**2 + a**2)
    return V_centrifugal + V_coulomb

def V_hydrogen_s_orbital(x):
    """예시 7: 1D 수소 s-오비탈 (l=0)""" 
    l = 0
    Z = 1      
    a = 0.5    
    V_coulomb = -2 * Z / np.sqrt(x**2 + a**2)
    V_centrifugal = 0.0
    return V_centrifugal + V_coulomb



# ============================================================
#  "직교" 기저 함수 (SHO Eigenfunctions)
# ============================================================
def h_basis(n, x):
    H_n = hermite(n)
    norm = 1.0 / np.sqrt((2**n) * factorial(n, exact=False) * np.sqrt(np.pi))
    return norm * H_n(x) * np.exp(-x**2 / 2.0)


# ============================================================
# 3️⃣ 홀/짝 분리 행렬 방정식 계산
# ============================================================
def solve_H_parity(N, V_func, parity):
    H = np.zeros((N, N))
    
    if parity == 'even':
        basis_indices = [2 * i for i in range(N)] 
    elif parity == 'odd':
        basis_indices = [2 * i + 1 for i in range(N)] 
    else:
        raise ValueError("Parity must be 'even' or 'odd'")
        
    basis = np.array([h_basis(idx, X_GRID) for idx in basis_indices])
    V_vals = V_func(X_GRID)
    V_sho_vals = V_sho(X_GRID)
    V_delta_vals = V_vals - V_sho_vals

    for i in range(N):
        n_i = basis_indices[i] 
        for j in range(i, N): 
            n_j = basis_indices[j]
            H_sho_ij = (n_i + 0.5) if (i == j) else 0.0
            integrand_V_delta = basis[i] * V_delta_vals * basis[j]
            V_delta_ij = np.trapezoid(integrand_V_delta, X_GRID)
            H[i, j] = H_sho_ij + V_delta_ij
            if i != j: H[j, i] = H[i, j] 

    try:
        evals, evecs = linalg.eigh(H)
        return H, evals[0], evecs[:, 0] 
    except linalg.LinAlgError:
        return None, np.nan, None


# ============================================================
# 4️⃣ 메인 분석 루프
# ============================================================
def run_stable_analysis(V_func, potential_name, N_max_run):
    
    print(f"\n=== Starting Stable Analysis for: {potential_name} (N_MAX={N_max_run}) ===")
    E_g_list, E_1_list = [], []

    for N in range(1, N_max_run + 1):
        print(f"▶ N = {N} (Using {N} even and {N} odd basis functions)")
        
        _, E_g, c_g_vec = solve_H_parity(N, V_func, 'even')
        _, E_1, c_1_vec = solve_H_parity(N, V_func, 'odd')
        
        if c_g_vec is None or c_1_vec is None:
             E_g, E_1 = np.nan, np.nan

        print(f"     E_g={E_g:.8f} (N_even={N}),      E_1={E_1:.8f} (N_odd={N})")
        E_g_list.append(E_g); E_1_list.append(E_1)

    # --- 에너지 수렴 그래프 ---
    Ns = range(1, N_max_run + 1)
    plt.figure(figsize=(10, 6))
    plt.plot(Ns, E_g_list, 'bo-', label='E_g (Even State)') 
    plt.plot(Ns, E_1_list, 'rs-', label='E_1 (Odd State)') 
    
    # [수정] SHO (검증용) 외의 모든 이론값 라인 제거
    if V_func == V_sho:
        plt.axhline(0.5, color='blue', linestyle=':', label='E_g (Theory=0.5)')
        plt.axhline(1.5, color='red', linestyle=':', label='E_1 (Theory=1.5)')
    
    plt.xlabel('N (Number of Basis Functions per Parity)')
    plt.ylabel('Energy')
    # [수정] 그래프 제목 단순화
    plt.title(f'{potential_name} - Energy Convergence')
    plt.grid(True); plt.legend()
    plt.show()

    print(f"\n--- Final Result for {potential_name} (N={N_max_run}) ---") 
    final_E_g = E_g_list[-1]
    final_E_1 = E_1_list[-1]
    final_delta_e = final_E_1 - final_E_g
    print(f"Final E_g = {final_E_g:.8f} (using {N_max_run} even basis)")
    print(f"Final E_1 = {final_E_1:.8f} (using {N_max_run} odd basis)")
    print(f"Energy Gap (Delta_E) = {final_delta_e:.8f}")
    
    # --- 파동함수 수식 및 그래프 준비 ---
    H_matrix_g, E_g_final, c_g_final = solve_H_parity(N_max_run, V_func, 'even')
    H_matrix_1, E_1_final, c_1_final = solve_H_parity(N_max_run, V_func, 'odd')
    
    # psi_1 (홀함수)의 부호 조정 (시각화 편의용)
    if c_1_final is not None and len(c_1_final) > 0 and c_1_final[0] < 0:
        print("[Note] Flipping the sign of psi_1 for preferred visualization (x>0 -> positive).")
        c_1_final = -c_1_final 

    basis_indices_g = [2 * i for i in range(N_max_run)]
    basis_indices_1 = [2 * i + 1 for i in range(N_max_run)]
    
    psi_g_str = f"psi_g(x) [N={N_max_run}] = "
    psi_1_str = f"psi_1(x) [N={N_max_run}] = "
    for i in range(N_max_run):
        if c_g_final is not None and abs(c_g_final[i]) > 1e-10:
            psi_g_str += f" ({c_g_final[i]:+.4f})*h_{basis_indices_g[i]}(x) "
        if c_1_final is not None and abs(c_1_final[i]) > 1e-10:
            psi_1_str += f" ({c_1_final[i]:+.4f})*h_{basis_indices_1[i]}(x) "
            
    print("\n--- Final Wavefunctions (Coefficients) ---")
    print(psi_g_str)
    print(psi_1_str)
    print(f"(h_n(x)는 n번째 SHO 고유 함수입니다.)")
    
    basis_g_final = np.array([h_basis(idx, X_GRID) for idx in basis_indices_g])
    basis_1_final = np.array([h_basis(idx, X_GRID) for idx in basis_indices_1])
    
    # None 체크 추가
    psi_g = np.dot(c_g_final, basis_g_final) if c_g_final is not None else np.zeros_like(X_GRID)
    psi_1 = np.dot(c_1_final, basis_1_final) if c_1_final is not None else np.zeros_like(X_GRID)
    
    V_plot = V_func(X_GRID)
    prob_g = psi_g**2
    prob_1 = psi_1**2

    # --- 그래프 로직 (l=1일 때와 아닐 때 분리) ---
    
    if V_func == V_hydrogen_p_orbital:
        # --- l=1 (p-오비탈) 특별 그래프: (V 줌인), (확률), (파형) 3개 ---
        fig, (ax_V, ax_prob, ax_psi) = plt.subplots(1, 3, figsize=(24, 7))
        # [수정] 그래프 제목 단순화
        fig.suptitle(f'{potential_name} - Wavefunctions Analysis', fontsize=16, y=1.02)

        # 1. 왼쪽: V(x) (줌인)
        ax_V.set_title('Potential V(x) [Zoomed-in]')
        ax_V.plot(X_GRID, V_plot, 'k--', label=f'V(x) ({potential_name})')
        ax_V.set_xlabel('x (Position)')
        ax_V.set_ylabel('Energy / V(x)')
        ax_V.set_ylim(-2, 2)
        ax_V.set_xlim(-X_LIMIT/2, X_LIMIT/2)
        ax_V.legend()
        ax_V.grid(True)
        
        # 2. 중간: 확률 밀도 |psi|^2
        ax_prob.set_title('Probability Density')
        ax_prob.plot(X_GRID, prob_g, 'b-', label=r'$|\psi_g(x)|^2$ (Even State)') 
        ax_prob.plot(X_GRID, prob_1, 'r-', label=r'$|\psi_1(x)|^2$ (Odd State - Physical 2p)') 
        ax_prob.set_xlabel('x (Position)')
        ax_prob.set_ylabel(r'Probability Density $|\psi(x)|^2$')
        ax_prob.legend()
        ax_prob.grid(True)
        ax_prob.set_xlim(-X_LIMIT/2, X_LIMIT/2)
        
        # 3. 오른쪽: 파동함수 psi
        ax_psi.set_title('Wavefunction Shape (psi)')
        ax_psi.plot(X_GRID, psi_g, 'b-', label=r'$\psi_g(x)$ (Even State)')
        ax_psi.plot(X_GRID, psi_1, 'r-', label=r'$\psi_1(x)$ (Odd State - Physical 2p)')
        ax_psi.axhline(0, color='k', linestyle='--', linewidth=0.5)
        ax_psi.set_xlabel('x (Position)')
        ax_psi.set_ylabel(r'Wavefunction Amplitude $\psi(x)$')
        ax_psi.legend()
        ax_psi.grid(True)
        ax_psi.set_xlim(-X_LIMIT/2, X_LIMIT/2)

    else:
        # --- 그 외 모든 (6개) 예시: 기존 2개 그래프 (twinx 사용) ---
        fig, (ax1, ax_psi) = plt.subplots(1, 2, figsize=(16, 6))
        # [수정] 그래프 제목 단순화
        fig.suptitle(f'{potential_name} - Wavefunctions Analysis', fontsize=16, y=1.02)

        # 1. 왼쪽: V(x)와 |psi|^2 (이중축)
        ax1.set_title('Potential vs Probability Density')
        color = 'k' 
        ax1.set_xlabel('x (Position)')
        ax1.set_ylabel('Energy / V(x)', color=color)
        ax1.plot(X_GRID, V_plot, color=color, linestyle='--', label=f'V(x) ({potential_name})')
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.grid(True) 

        ax2 = ax1.twinx() 
        color = 'tab:blue' 
        ax2.set_ylabel(r'Probability Density $|\psi(x)|^2$', color=color) 
        ax2.plot(X_GRID, prob_g, 'b-', label=r'$|\psi_g(x)|^2$')
        ax2.plot(X_GRID, prob_1, 'r-', label=r'$|\psi_1(x)|^2$')
        ax2.tick_params(axis='y', labelcolor=color)
        
        max_prob = max(np.nanmax(prob_g), np.nanmax(prob_1))
        if not np.isfinite(max_prob): max_prob = 1.0
        ax2.set_ylim(-0.1, max_prob * 1.2) 

        lines, labels = ax1.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels() 
        ax2.legend(lines + lines2, labels + labels2, loc='upper center')

        V_min = np.nanmin(V_plot[np.abs(X_GRID) < 5])
        V_max = np.nanmax(V_plot[np.abs(X_GRID) < 5])
        if not np.isfinite(V_min): V_min = -1.0
        if not np.isfinite(V_max): V_max = E_1_final * 10
        
        if np.abs(V_max - V_min) < 1e-6:
            V_min -= 1.0
            V_max += 1.0

        ax1.set_ylim(V_min - abs(E_1_final-E_g_final)*0.5, V_max + abs(E_1_final-E_g_final)*1.5)
        ax1.set_xlim(-X_LIMIT/2, X_LIMIT/2)
        
        # 2. 오른쪽: 파동함수(psi) 자체의 파형
        ax_psi.set_title('Wavefunction Shape (psi)')
        # 레이블 수정 (짝/홀 명시)
        ax_psi.plot(X_GRID, psi_g, 'b-', label=r'$\psi_g(x)$ (Lowest Even State)') 
        ax_psi.plot(X_GRID, psi_1, 'r-', label=r'$\psi_1(x)$ (Lowest Odd State)') 
        ax_psi.axhline(0, color='k', linestyle='--', linewidth=0.5)
        ax_psi.set_xlabel('x (Position)')
        ax_psi.set_ylabel(r'Wavefunction Amplitude $\psi(x)$')
        ax_psi.legend()
        ax_psi.grid(True)
        ax_psi.set_xlim(-X_LIMIT/2, X_LIMIT/2)

    fig.tight_layout() 
    plt.show()

    # --- 에너지 분산(Variance) 계산 ---
    if H_matrix_g is None or H_matrix_1 is None:
        print("--- Quality Check Skipped (Matrix Error) ---")
        return

    print(f"\n--- Quality Check: Energy Variance (N={N_max_run}) ---")
    
    # None 체크 추가
    variance_g = np.nan
    if c_g_final is not None:
        H_g_squared = H_matrix_g @ H_matrix_g
        H2_exp_g = c_g_final.T @ H_g_squared @ c_g_final
        variance_g = H2_exp_g - (E_g_final**2)
    
    print(f"[Lowest Even State (E_g)]") # 레이블 수정
    print(f"  <H>_g            = {E_g_final:.10f}")
    if c_g_final is not None:
        print(f"  <H^2>_g          = {H2_exp_g:.10f}")
    print(f"  Variance_g (0 is best) = {variance_g:.6e}")

    # None 체크 추가
    variance_1 = np.nan
    if c_1_final is not None:
        H_1_squared = H_matrix_1 @ H_matrix_1
        H2_exp_1 = c_1_final.T @ H_1_squared @ c_1_final
        variance_1 = H2_exp_1 - (E_1_final**2) 
    
    print(f"\n[Lowest Odd State (E_1)]") # 레이블 수정
    print(f"  <H>_1            = {E_1_final:.10f}")
    if c_1_final is not None:
       print(f"  <H^2>_1          = {H2_exp_1:.10f}")
    print(f"  Variance_1 (0 is best) = {variance_1:.6e}")
    
    if V_func == V_sho:
        print(f"  (Note: For SHO, variance must be 0. Non-zero value is numerical error.)")
        
    # --- <|x|> (1D 보어 반지름) 계산 ---
    # 수소 원자 예시 2개에 대해서만 계산 (None 체크 추가)
    if (V_func == V_hydrogen_s_orbital or V_func == V_hydrogen_p_orbital) and \
       (c_g_final is not None and c_1_final is not None) :
        
        print(f"\n--- Quality Check: 1D Expectation Value <|x|> ---")
        
        # 1. s-오비탈 (l=0)
        if V_func == V_hydrogen_s_orbital:
            # 물리적인 해는 '짝수' 바닥 상태 (psi_g)
            integrand_x_g = prob_g * np.abs(X_GRID)
            exp_x_g = np.trapezoid(integrand_x_g, X_GRID)
            print(f"  [s-orbital (l=0)] Physical State is E_g (Even)")
            print(f"  Calculated <|x|> = {exp_x_g:.6f}")
            print(f"  (Ref: 3D Theory <r>_1s = 1.5)") 

        # 2. p-오비탈 (l=1)
        if V_func == V_hydrogen_p_orbital:
            # 물리적인 해는 '홀수' 바닥 상태 (psi_1)
            integrand_x_1 = prob_1 * np.abs(X_GRID)
            exp_x_1 = np.trapezoid(integrand_x_1, X_GRID)
            print(f"  [p-orbital (l=1)] Physical State is E_1 (Odd)")
            print(f"  Calculated <|x|> = {exp_x_1:.6f}")
            print(f"  (Ref: 3D Theory <r>_2p = 5.0)")


    print(f"--- {potential_name} Analysis Complete ---")


# ============================================================
# 5️⃣ 실행 (총 7개 예시, N_max 개별 설정)
# ============================================================
if __name__ == "__main__":
    
    # 1-4번: 간단한 문제 (N=10이면 충분)
    run_stable_analysis(V_sho, "Harmonic Oscillator (SHO)", N_max_run=10)
    run_stable_analysis(V_anharmonic, "Anharmonic Oscillator", N_max_run=10)
    run_stable_analysis(V_double_well, "Double-Well Potential", N_max_run=10)
    run_stable_analysis(V_finite_well, "Finite Square Well", N_max_run=10)
    
    # [수정] 5-7번 이름 단순화
    
    # 5번: soft 쿨롱 (N=15)
    run_stable_analysis(V_soft_coulomb_toy, "1D Soft-Coulomb (V=-1/r type)", N_max_run=15)
    
    # 6번: 1D 수소 p-오비탈 (N=15)
    run_stable_analysis(V_hydrogen_p_orbital, "1D Hydrogen p-orbital (l=1 model)", N_max_run=15)

    # 7번: 1D 수소 s-오비탈 (N=15)
    run_stable_analysis(V_hydrogen_s_orbital, "1D Hydrogen s-orbital (l=0 model)", N_max_run=15)

    print("\nAll 7 analyses complete ✅")
```
